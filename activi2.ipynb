```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df0 = pd.read_csv('pin.txt', skiprows=4, sep='\s+')
```

```python
df0.head()
```

```python
df = pd.DataFrame(df0)
```

```python
df.dtypes
```

```python


df['FECHA'] = pd.to_datetime(df.apply(lambda x: x['DD/MM/AAAA'] + ' ' + x['HH:MM'], 1), dayfirst=True)
df = df.drop(['DD/MM/AAAA', 'HH:MM'], 1)

```

```python
df.head()
```

```python
df.describe()
```

```python
plt.figure(); df.VELS.plot(); plt.legend(loc='best')
plt.title("Variacin de la Rapidez de los Vientos")
plt.ylabel("Rapidez (m/s)")
plt.grid(True)
plt.show()
```

```python
df3 = df[['VELS','VELR']]
plt.plot(df.FECHA,df3); plt.legend(["Rapidez del viento","Rapidez de las rfagas"])
plt.title("Rapidez del Viento y Rafagas")
plt.xlabel("Fecha y Hora")
plt.ylabel("Rapidez(m/s)")
plt.grid(True)
plt.show()
```

```python
plt.plot(df.FECHA,df['RADSOL'])
plt.title("Radiacin solar")
plt.xlabel("Fecha y Hora")
plt.ylabel("Radiacin (W/m2)")
plt.grid(True)
plt.show()
```

```python
plt.plot(df.FECHA,df.TEMP)
plt.title("Variacin de Temperatura.")
plt.ylabel("Temp C")
plt.grid(True)
plt.show()
plt.plot(df.FECHA,df.HR)
plt.title("Variacin de Humedad Relativa.")
plt.ylabel("HR (%)")
plt.grid(True)
plt.show()
```

```python
df4 = df[['TEMP','HR']]
plt.figure(); df4.plot(); plt.legend(loc='best')
plt.title("Variacin de Temperatura y Humedad Relativa.")
plt.ylabel("Temp C /(%) HR")
plt.grid(True)
plt.show()
```
